```python
import numpy as np
from random import randint
url = 'https://raw.githubusercontent.com/sharmaroshan/Clustering-of-Mall-Customers/master/Mall_Customers.csv'
import matplotlib.pyplot as plt
data = np.loadtxt(url, delimiter=',', dtype=object, skiprows=1)
data[:, 1] = data[:, 1] == 'Male'
data = data.astype(int)
data=data[:,1:]
#print(data)
def distE(point1,point2):
    x=(point1-point2)**2
    return (sum(x))**(1/2)
def kmeans(obs, k, iter = 40, thresh = 1e-05):
    al=obs.size
    args=obs[0].size
    dat=al//args
    centre=obs[:k].copy()
    disp=np.ones((k,args))
    own=np.ones(dat,dtype=int)
    print(own.dtype)
    for z in range(iter):
        clon=disp.copy()
        for i in range(dat):
            s=10000
            for j in range(k):
                if distE(obs[i],centre[j]) <s:
                    s=distE(obs[i],centre[j])
                    own[i]=j
        for j in range(k):
            mask=own==j
            centre[j]=np.mean(obs[mask],axis=0)
            disp[j]=np.std(obs[mask],axis=0)
        if np.sum((clon-disp)**2)<thresh:
            break
    print(np.mean(disp,axis=0))
    colors = ['blue','red','green','yellow','goldenrod']
    col= []
    for i in range(dat):
            col.append(colors[own[i]])   
    return np.column_stack([obs,own])#добавляю цвета к матрице
x=kmeans(data,5)
print(x)
```

```python
colors = ['blue','red','green','yellow','goldenrod']
import matplotlib.ticker as ticker
col= []
for i in range(len(x)):
            col.append(colors[x[i,4]]) 
plt.scatter(x[:, 1], x[:, 2],c=col)
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.show()
#sorted(x, key=lambda row: int(row[1]))
#x=x[x[:,4].argsort()]
#x=x[np.argsort(x[:, 1])]
#print(x)
#print(np.sort(x.view('i8 i8 i8 i8 i8'),order=['f1'],axis=0))
```
