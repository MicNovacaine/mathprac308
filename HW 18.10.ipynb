```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
```

```python
plt.xticks(data[:,0].astype(int)[::2])
plt.plot(data[:,0], data[:,1],c='r')
plt.plot(data[:,0], data[:,2],c='b')
plt.plot(data[:,0], data[:,3],c='g')
plt.legend(('Hares', 'Lynxes','Carrots'), loc='upper right')
plt.show()
```

Task 2

```python
fig, ax = plt.subplots()

ax.bar(data[:,0], data[:,1], 0.5, label='Hares')
ax.bar(data[:,0], data[:,2], 0.5, bottom=data[:,1],label='Lynxes')

ax.set_ylabel('Years')
ax.set_title('Hares/Lynxes')
ax.legend()

plt.show()
```

Task 3

Task 7 
Find the top 2 years for each species when they had the lowest populations.

```python
a0=data[:,0]
a1=data[:,1]
a2=data[:,2]
a1=a1.argsort()[:2]
print(a1)
a2=a2.argsort()[:2]
print(a2)
plt.xticks(data[:,0].astype(int)[::2])
plt.plot(data[:,0], data[:,1],c='r')
plt.scatter(data[a1,0], data[a1,1], c='y')
plt.plot(data[:,0], data[:,2],c='b')
plt.scatter(data[a2,0], data[a2,2], c='g')
plt.legend(('Hares', 'Lynxes'), loc='upper right')
plt.show()
```

Task 8 
Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

```python
a1=data[:,1]
a11=np.delete(a1,0)
a11=np.append(a11,np.mean(a1))
a=a1-a11
print(np.corrcoef(a,data[:,2]))
plt.xticks(data[:,0].astype(int)[::2])
plt.plot(data[:,0], a,c='r')
plt.plot(data[:,0], data[:,2],c='b')
plt.legend(('Change in hare population ', 'Lynxes'), loc='upper right')
plt.show()
```

Task 9 Show population of hares vs carrots and hares vs lynxes.


```python
hares = np.argsort(data[:, 1])
plt.figure(figsize=(10,10))
plt.subplot(2, 1, 1)
plt.plot(data[hares, 1], data[hares, 2], c='y')
plt.title('Hares vs Lynxes')
plt.xlabel('H')
plt.ylabel('L')
plt.subplot(2, 1, 2)
plt.plot(data[hares, 1], data[hares, 3], c= 'black')
plt.title('Hares vs Carrots')
plt.xlabel('H')
plt.ylabel('C')
plt.show()
```
